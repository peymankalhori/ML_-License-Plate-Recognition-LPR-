```python
#تمامی تغییرات رو میتونیم داخل یم حلقه برای تمامی عکس های دیتاست انجام بدیم
files = os.listdir('1')

#ye array ba 256 vurudi dorost kardim ke x hammuno ja bedim tush
x = np.empty((0,256))
#ye array khali ham dorost kardim ke begim har x ei ke khundi y esh mishe masalan felan adad!!
y = np.array([])

for file_name in files:
    im = cv2.imread('1/' + file_name)
    im2 = cv2.resize(im,(8,32))
    im3 = cv2.cvtColor(im2,cv2.COLOR_BGR2GRAY)
    im4 = im3.reshape(256)
    x=np.append(x,[im4],axis=0)
    y=np.append(y,'1')
```
